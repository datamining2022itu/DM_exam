```python
# Finalised mobilities started in 2014 - KA1 - src https://data.europa.eu/data/datasets/erasmus-mobility-statistics-2014-2018?locale=en

url = 'https://data.europa.eu/euodp/data/storage/f/2020-08-11T140550/Finalised%20mobilities%20started%20in%202018%20-%20KA1.csv'

import pandas as pd
```

```python
df = pd.read_csv(url, on_bad_lines='skip', sep=';')
```

```python
partner_countries = ['PT', 'AT', 'ES', 'UA',  
        'BE', 'IT',  
         'BG', 
       'CZ', 'DE',
       'HR',  
         'DK',  'FI', 'EE',
       'NO', 'FR', 'EL','IE',  
       'MT', 'LV','RO', 'SI', 'HU', 
       'LI', 'LT', 'PL', 'LU' 'SK', 
       'TR', 'MK', 'NL', 'SE']
#UA ukraine
# EL greece
#AT austria
# RS is serbia
# SERBIA MACEDONIA TURKEY AND LICHSTENEIN NORWAY
non_partner = ['IL', 'CN', 'KR', 'RS','ID','US','NZ','CU','LK','BR', 'JP','ME', 'AL',
               'BA', 'TT', 'UG', 'CR', 'NI', 'XK', 'AM', 'NP', 'IN',  'DZ', 'TN',
               'VN', 'UZ', 'MD', 'KZ', 'MA', 'LB','KH', 'BY', 'BF', 'JO',
               'TZ', 'HK', 'CL', 'CY', 'MN', 'SG', 'MY', 'EG', 'EC', 'ZM','AZ', 'DO', 'AR', 'FJ', 
               'GE', 'PG','MX', 'KG', 'CO', 'PK', 'AF', 'TJ', 'KE', 'MM', 'CM',
               'BJ', 'PH', 'BT', 'UK', 'BD', 'GH', 'TW', 'BN', 'MU', 'HT', 'SY', 'LY',
               'TG', 'PY', 'BZ', 'JM', 'CV', 'SN', 'MZ', 'MR', 'TM', 'GT', 'BO', 'UY', 'PE',
                'BB', 'CI', 'ML', 'BW', 'SV', 'IS', 'MO', 'GM', 'LA', 
                'SR', 'HN', 'TL', 'SD', 'LS', 'CF', 'MW', 'NG', 'IR', 'RW', 'CD', 
               'MG', 'NE', 'GA', 'SC', 'AO', 'IQ', 'GW', 'PA', 'AW', 'VE', 'ZW',
               'MS', 'SX','CH', 'CW', 'GS' ]
#XK kosovo
```

```python
df = df.rename(columns={"Mobility Start Month": "startMonth", "Mobility End Month": "endMonth",
                   'Mobility Duration': 'duration',
                   'Activity (mob)': 'activity', 'Field of Education': 'educationField',
                   'Participant Nationality': 'participantNationality',
                   'Participant Gender': 'participantGender',
                   'Participant Age': 'participantAge',
                   'Sending Country Code': 'sendingCountry',
                   'Sending City': 'sendingCity',
                   'Sending Organization': 'sendingOrganization',
                   'Receiving Country Code': 'receivingCountry',
                   'Receiving Organization': 'receivingOrganization',
                   'Participants' : 'participants'
                   })
#you can only choose country in erasmus, not city. 
# participants drop everything that is more than 1.
```

```python
df = df[["startMonth", 
    "endMonth",
    'duration',
    'activity', 
    'educationField',
    'participantNationality',
    'participantGender',
    'participantAge',
    'sendingCountry',
    'sendingCity',
    'sendingOrganization',
    'receivingCountry',
    'receivingOrganization',
    'participants']]
```

```python
# value counts for non-partner sending countries
df[df['sendingCountry'].isin(non_partner)]['sendingCountry'].value_counts()
```

```python
# value counts for partner sending countries
df[df['sendingCountry'].isin(partner_countries)]['sendingCountry'].value_counts()
```
